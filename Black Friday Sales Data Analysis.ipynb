## Combining age and marital status

```python
for i in range(len(df)):
    print(df["Gender"][i] +"_"+ str(df["Marital_Status"][i]))
    break
```

```python
l=[]
for i in range(len(df)):
    l.append(df["Gender"][i] +"_"+ str(df["Marital_Status"][i]))
    
```

```python
df["Marital_Gender"]=l
```
